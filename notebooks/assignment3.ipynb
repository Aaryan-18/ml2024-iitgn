```python
import numpy as np
np.random.seed(0)
```

```python
x = np.linspace(0, 10, 30)
```

```python
y = np.sin(x) + 10*x**2 - 3*x**3 + 0.2*x**4 + 15*np.random.randn(*x.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
deg = 6
poly = PolynomialFeatures(deg, include_bias=False)
```

```python
lr = LinearRegression()
lr.fit(poly.fit_transform(x.reshape(-1, 1)), y)
```

```python
plt.plot(x, lr.predict(poly.fit_transform(x.reshape(-1, 1))),color='k',label='Fit (degree: {})'.format(deg))
plt.scatter(x, y)
lr.coef_
```
